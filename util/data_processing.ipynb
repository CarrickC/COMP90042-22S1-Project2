```python
data_id_f = open('data/train/train.data.txt', 'r')
data_label = open('data/train/train.label.txt', 'r')

databank = {'text':[], 'label':[], 'source_id':[]}
for i, (line, label) in enumerate(zip(data_id_f.readlines(), data_label.readlines())):
    if label == 'rumour\n':
        label = 1
    else:
        label = 0
    ids = line.strip('\n').split(',')
    id = ids[0]
    if os.path.isfile('data/train/tweet_objects/'+id+'.json'):
        with open('data/train/tweet_objects/'+id+'.json', 'r') as f:
            data = json.load(f)
            source = re.sub(r'https?://t.co/[a-zA-Z0-9]+', '', data['text'].strip('\n'))
    else:
        source = '""'
    databank['text'].append(source)
    databank['label'].append(label)
    databank['source_id'].append(i)
    if len(ids) > 1:
        for id in ids[1:]:
            if os.path.isfile('data/train/tweet_objects/'+id+'.json'):
                with open('data/train/tweet_objects/'+id+'.json', 'r') as f:
                    data = json.load(f)
                    retweet = re.sub(r'https?://t.co/[a-zA-Z0-9]+', '', data['text'].strip('\n'))
            else:
                retweet = '""'
            databank['text'].append(source + ' ' + retweet)
            databank['label'].append(label)
            databank['source_id'].append(i)
data_label.close()
data_id_f.close()
```
